# Insert Datos

- InsertMany

```python
from pymongo import MongoClient
from dotenv import load_dotenv
import os

# Cargar variables de entorno
load_dotenv()

# Obtener configuración
MONGODB_URI_ATLAS = os.getenv("MONGODB_URI_ATLAS")
DATABASE_NAME = os.getenv("MONGODB_Data")

# Crear cliente MongoDB
client = MongoClient(
    MONGODB_URI_ATLAS,
    tls=True,
    tlsAllowInvalidCertificates=True
)

# Conectar a la base de datos
db = client[DATABASE_NAME]
libros = db.libros

libro = {
    'titulo': 'La casa verde',
    'autor': 'Mario Vargas Llosa',
    'anio': 1967,
    'categorias': [
        'novela', 'ficcion'
    ],
    'stock': 4
}

#Insert One
res = libros.insert_one(libro)
print("InsertOne id: ", res.inserted_id)
```

```python
#Inset Many (varias inserciones)
docs = [
    #Libro 2
    {
        'titulo': 'Cien años de soledad',
        'autor': 'Gabriel Garcia Marquez',
        'anio': 1967,
        'categorias': [
            'novela', 'ficcion magica'
        ],
        'stock': 6
    },
    #Libro 3
    {
        'titulo': 'Don Quijote de la Mancha',
        'autor': 'Miguel de Cervantes',
        'anio': 1605,
        'categorias': [
            'novela', 'aventuras'
        ],
        'stock': 3
    }
]

res_many = libros.insert_many(docs)
print("InsertMany ids: ", res_many.inserted_ids)
```

```python
# Filtros con where
for doc in libros.find(
    {'autor' : 'Gabriel Garcia Marquez'}
):
    print(doc['titulo'], doc['anio'])
```

```python
#Comparar
#Publicaciones mayores año 1960

for doc in libros.find({'anio': {'$gt': 1960}}):
    print(doc['titulo'], doc['anio'])
```
